```python
import pandas as pd
import seaborn as sns
import folium
import json
sns.set_theme(style="ticks", palette="pastel")

dataset = pd.read_csv('policelog_ba.txt', sep=';')
accidents = pd.read_csv('accidentlog_ba.txt', sep=';')
#dataset["timestamp"] = dataset["timestamp"].div(10000000)
dataset["datetime"] = pd.to_datetime(dataset["timestamp"],unit='ms')
dataset
```

```python
mapt = folium.Map(location=["48.10","17.17"], zoom_start=12, tiles="OpenStreetMap")

for index, row in dataset.iterrows():
    tmp_arr = json.loads(row[1])
    for coord in tmp_arr:
        #print(coord['long'])
        folium.CircleMarker(
        [coord['lat'],coord['long']],
        fill=True,
        color="blue",
        fill_color="blue",
        fill_opacity=1,
        radius=2,
        ).add_to(mapt)
        
for index, row in accidents.iterrows():
    tmp_arr = json.loads(row[1])
    for coord in tmp_arr:
        #print(coord['long'])
        folium.CircleMarker(
        [coord['lat'],coord['long']],
        fill=True,
        color="red",
        fill_color="red",
        fill_opacity=1,
        radius=4,
        ).add_to(mapt)
       
mapt
    
```
