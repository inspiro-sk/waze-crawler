```python
import json
import pandas as pd
import seaborn as sns
import folium
```

```python
df = pd.read_csv("police_coordinates.log", sep="\t", names=['timestamp', 'coords'], index_col=False)
df.reset_index(inplace=False)
df.head()
```

```python
mapt = folium.Map(location=["49.05", "19.30"],
                  zoom_start=10, tiles="OpenStreetMap")

for index, row in df.iterrows():
    tmp_arr = json.loads(row[1])
    for coord in tmp_arr:
        # print(coord['lat'], coord['long'], coord['type'])
        if coord['type'] == 'POLICE':
            folium.CircleMarker(
                [coord['lat'], coord['long']],
                fill=True,
                color="blue",
                fill_color="blue",
                fill_opacity=1,
                radius=2,
            ).add_to(mapt)
        else:
            # accidents markers
            folium.CircleMarker(
                [coord['lat'], coord['long']],
                fill=True,
                color="red",
                fill_color="red",
                fill_opacity=1,
                radius=2,
            ).add_to(mapt)


mapt
```
